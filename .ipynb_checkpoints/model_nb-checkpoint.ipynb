# Emotion Recognizer

## Importing libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

## Loading data

```python
train_data = open(r"train.txt").readlines()
texts = []
labels = []

for i in train_data:
    j = i.split(";")
    texts.append(j[0])
    labels.append(j[1].rstrip("\n"))
texts = np.array(texts)
labels = np.array(labels)
```

## Encoding the labels

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
labels = encoder.fit_transform(labels)
labels
```

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)
cleaned_texts = [remove_stopwords(word) for word in texts]
```

## Tokenization and Padding

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(cleaned_texts).toarray()
```

```python
model_nb = MultinomialNB()
model_nb.fit(X,labels)
```

```python
test_data = open(r"val.txt").readlines()
test_texts = []
test_labels = []

for i in test_data:
    j = i.split(";")
    test_texts.append(j[0])
    test_labels.append(j[1].rstrip("\n"))
test_texts = np.array(test_texts)
test_labels = np.array(test_labels)
test_labels = encoder.fit_transform(test_labels)
cleaned_test_texts = [remove_stopwords(word) for word in test_texts]
X_test = vectorizer.transform(cleaned_test_texts).toarray()
```

```python
predictions = model_nb.predict(X_test)
```
